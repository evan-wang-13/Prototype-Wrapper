```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load data from a pickle file
def load_data_from_pickle(file_path):
    with open(file_path, 'rb') as file:
        data = pickle.load(file)
    return data

# Function to plot the data
def visualize_data(data):
    if isinstance(data, pd.DataFrame):
        # If the data is a DataFrame, plot a pairplot or histogram
        sns.pairplot(data)
        plt.show()
    elif isinstance(data, list) and all(isinstance(x, (int, float)) for x in data):
        # If the data is a list of numbers, plot a simple line plot
        plt.plot(data)
        plt.title('Line Plot of Data')
        plt.xlabel('Index')
        plt.ylabel('Value')
        plt.show()
    else:
        print("Data type not supported for visualization.")

# Example usage
file_path = 'data/real_actions_PPO.pkl'  # Replace with your .pkl file path
data = load_data_from_pickle(file_path)
```
